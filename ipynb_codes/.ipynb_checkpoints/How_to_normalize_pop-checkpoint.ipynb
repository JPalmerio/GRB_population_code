```python
import sys
src_dir = '/Users/palmerio/Science_projects/GRB_population_model/src/'
sys.path.insert(0, src_dir)
from GRB_population import GRBPopulation, create_GRB_population_from
import plotting_functions as pf
import miscellaneous as msc
import io_grb_pop as io
from cosmology import init_cosmology
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('paper')
```

```python
paths_to_dir, paths_to_files = io.generate_paths()
cosmo = init_cosmology(paths_to_dir['cosmo'])

config, params, instruments, samples, obs_constraints = io.read_init_files(paths_to_files)
config['Nb_GRBs'] = '1e5'
config['samples'] = ['Stern', 'EpGBM', 'eBAT6', 'SHOALS']
config['output_dir'] = 'how_to_normalize_GRB_pop'
paths_to_dir['output'] = paths_to_dir['output']/config['output_dir']

incl_samples, incl_instruments, incl_constraints = io.create_config(config,
                                                                    samples,
                                                                    instruments,
                                                                    obs_constraints)
```

```python
params['luminosity_function']['logLmin'] = 49.7
gp1 = create_GRB_population_from(cosmo, params, incl_instruments, incl_samples, incl_constraints)
```

```python
from constants import T_live_BATSE

# Look at raw histogram of number counts as a function of pflx_BATSE
fig, axes = plt.subplots(2,1, figsize=(10,15), sharex=True)

def plot_cstr(ax, GRB_pop, hist_key='hist', err_key='err', name='Stern', div_dbin=False, div_norm=None,
              y_log=True, constraints=None, **kwargs):
    # Unnormalized hist
    bins = GRB_pop.mock_constraints[name]['bins']
    err = GRB_pop.mock_constraints[name][err_key]
    hist = GRB_pop.mock_constraints[name][hist_key]
    if div_dbin:
        delta_bin = np.log10(bins[1:]/bins[:-1])
        hist = hist / delta_bin
        err = err / delta_bin
    if div_norm is not None:
        hist = hist / div_norm
        err = err / div_norm
    if y_log:
        y_mod, y_mod_errp, y_mod_errm = msc.lin_to_log(hist, err)
    else:
        y_mod = hist
        y_mod_errp = err
        y_mod_errm = err

    # Make it look better for plots
    x_plottable, y_mod_plottable = pf.plottable_hist(bins[:-1], y_mod, last_bin_edge=bins[-1])
    y_mod_errm_plottable = pf.double_array(y_mod_errm)
    y_mod_errp_plottable = pf.double_array(y_mod_errp)

    # Plot them
    art, = ax.plot(x_plottable, y_mod_plottable, **kwargs)
    ax.fill_between(x_plottable,
                    y_mod_plottable+y_mod_errp_plottable,
                    y_mod_plottable-y_mod_errm_plottable,
                    alpha=0.6, color=plt.getp(art,'color'), zorder=plt.getp(art,'zorder'))
    if constraints is not None:
        # Observations
        obs_bins = constraints[name]['bins']
        obs_err = constraints[name]['err']
        obs_hist = constraints[name]['hist']
        if div_dbin:
            obs_delta_bin = np.log10(obs_bins[1:]/obs_bins[:-1])
            obs_hist = obs_hist / obs_delta_bin
            obs_err = obs_err / obs_delta_bin
        if div_norm is not None:
            obs_hist = obs_hist / div_norm
            obs_err = obs_err / div_norm
        if y_log:
            obs_hist, obs_err, _ = msc.lin_to_log(obs_hist, obs_err)
        x, x_errp, x_errm = pf.xerr_from_bins(obs_bins, logscale=True)
        ax.errorbar(x, obs_hist, xerr=[x_errm, x_errp], yerr=obs_err, color='k', fmt='none', label='Observed')
    return

plot_cstr(ax=axes[0], GRB_pop=gp1, hist_key='hist_unnormed', err_key='err_unnormed', constraints=incl_constraints, label='Raw')
plot_cstr(ax=axes[0], GRB_pop=gp1, hist_key='hist', err_key='err',
          label=r'Normalized by $A$ ='+' {:.3e}'.format(gp1.mock_constraints['Stern']['norm']))

axes[0].legend()
axes[0].set_xscale('log')
axes[0].set_ylabel(r'log $\Delta N$ in 4 pi')

# Now look at the logRlogN
plot_cstr(ax=axes[1], GRB_pop=gp1, hist_key='hist_unnormed', err_key='err_unnormed',
          div_dbin=True,
          label='Raw')

plot_cstr(ax=axes[1], GRB_pop=gp1, hist_key='hist', err_key='err',
          div_dbin=True, div_norm=T_live_BATSE, constraints=incl_constraints,
          label=r'Normalized by $A/T_{\rm live}^{\rm BATSE}$ ='+' {:.3e} yr-1'.format(gp1.mock_constraints['Stern']['norm']/T_live_BATSE))
plot_cstr(ax=axes[1], GRB_pop=gp1, hist_key='hist_unnormed', err_key='err_unnormed',
          div_dbin=True, div_norm=gp1.normalization['T_sim'],
          label=r'Normalized by 1/$T_{\rm sim}$ ='+' {:.3e} yr-1'.format(1./gp1.normalization['T_sim']))


axes[1].legend()
axes[1].set_xlabel(r'$N^{\rm pk}_{50-300\,\rm keV}~\rm{[ph\,cm^{-2}\,s^{-1}]}$')
axes[1].set_ylabel(r'log $\Delta R/\Delta \,(\log\,N^{\rm pk})$')


plt.show()
```
