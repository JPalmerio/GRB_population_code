```python
def compare_redshift_to_SHOALS(population, plot_obs=True, ax=None, **kwargs):
    SHOALS_file = '../../Frederic/catalogs/SHOALS_cat/SHOALS_cat.txt'
    df_SHOALS_obs = pd.read_csv(SHOALS_file, sep='\t', header=3, low_memory=False)
    keys = ['S_BAT', 'z']
    for key in keys:
        df_SHOALS_obs[key] = pd.to_numeric(df_SHOALS_obs[key], errors='coerce')
    df_SHOALS_obs = df_SHOALS_obs.dropna()

    z_bins = np.asarray([0., 1.5, 2.5, 4.5, 6.5, 10])
    if ax is None:
        fig, ax = plt.subplots(figsize=(8,6))
    if plot_obs:
        hist_obs, z_bins, _ = ax.hist(df_SHOALS_obs['z'], bins=z_bins, edgecolor='k', color='brown', alpha=0.6,
                                      label='SHOALS obs (N={})'.format(len(df_SHOALS_obs)), density=True)
        sns.kdeplot(df_SHOALS_obs['z'], ax=ax, color='brown', label=None)

    ax.set_xlim(0,8)
    ax.set_xlabel('Redshift (z)')
    ax.set_title('SHOALS redshift distribution')
    return z_bins

def plot_SHOALS_obs(ax=None, kde=True, bins=None, kde_kwargs={}, **kwargs):
    SHOALS_file = '../../Frederic/catalogs/SHOALS_cat/SHOALS_cat.txt'
    df_SHOALS_obs = pd.read_csv(SHOALS_file, sep='\t', header=3, low_memory=False)
    keys = ['S_BAT', 'z']
    for key in keys:
        df_SHOALS_obs[key] = pd.to_numeric(df_SHOALS_obs[key], errors='coerce')
    df_SHOALS_obs = df_SHOALS_obs.dropna()
    
    if bins is None:
        z_bins = np.asarray([0., 1.5, 2.5, 4.5, 6.5, 10])
    else:
        z_bins = bins
    if ax is None:
        fig, ax = plt.subplots(figsize=(8,6))
    
    hist_obs, z_bins, _ = ax.hist(df_SHOALS_obs['z'], bins=z_bins, **kwargs)
    if kde:
        sns.kdeplot(df_SHOALS_obs['z'], ax=ax, **kde_kwargs)

    ax.set_xlim(0,8)
    ax.set_xlabel('Redshift (z)')
    ax.set_title('SHOALS redshift distribution')
    return z_bins
```

```python
def plot_eBAT6_obs(ax=None, kde=True, bins=None, kde_kwargs={}, **kwargs):
    eBAT6_file = '../../Frederic/catalogs/BAT6_cat/eBAT6_cat.txt'
    df_eBAT6_obs_tot = pd.read_csv(eBAT6_file, sep='|', header=3, low_memory=False)
    df_eBAT6_obs_tot.rename(columns=lambda x:x.strip(), inplace=True)
    
    df_eBAT6_obs = df_eBAT6_obs_tot['redshift']
    df_eBAT6_obs = pd.to_numeric(df_eBAT6_obs, errors='coerce')
    df_eBAT6_obs = df_eBAT6_obs.dropna()
    
    if bins is None:
        z_bins = np.asarray([0., 1.5, 2.5, 4.5, 6.5, 10])
    else:
        z_bins = bins
    if ax is None:
        fig, ax = plt.subplots(figsize=(8,6))
    
    hist_obs, z_bins, _ = ax.hist(df_eBAT6_obs, bins=z_bins, **kwargs)
    if kde:
        sns.kdeplot(df_eBAT6_obs, ax=ax, **kde_kwargs)

    ax.set_xlim(0,8)
    ax.set_xlabel('Redshift (z)')
    ax.set_title('eBAT6 redshift distribution')
    return z_bins
```

```python
def z_selection_func(pop, samp_num, samp_denum, bins=None, ax=None, **kwargs):
    if ax is None:
        fig, ax = plt.subplots()
    if bins is None:
        bins = np.arange(0.,10., 0.4)
    
    hist_num, bins = np.histogram(pop[(pop[f'pdet_{samp_num}'] == 1)]['z'], bins=bins)
    hist_denum, bins = np.histogram(pop[(pop[f'pdet_{samp_denum}'] == 1)]['z'], bins=bins)
    frac = hist_num / hist_denum
    bins_mid = (bins[1:]+bins[:-1]) / 2.
    ax.plot(bins_mid, frac, **kwargs)
    return
```
